# San Francisco Rental Prices Dashboard

In this notebook, you will compile the visualizations from the previous analysis into functions that can be used for a Panel dashboard.

```python
# imports
import panel as pn
from panel.interact import interact
import plotly.express as px
import pandas as pd
import hvplot.pandas
import matplotlib.pyplot as plt
import os
from pathlib import Path
from dotenv import load_dotenv
pn.extension('plotly')
```

# Import Data

```python
# Import the CSVs to Pandas DataFrames
file_path = Path("Data/sfo_neighborhoods_census_data.csv")
sfo_data = pd.read_csv(file_path, index_col="year")
sfo_data.dropna(inplace=True)

file_path = Path("Data/neighborhoods_coordinates.csv")
df_neighborhood_locations = pd.read_csv(file_path)
```

## Panel Visualizations

In this section, you will copy the code for each plot type from your analysis notebook and place it into separate functions that Panel can use to create panes for the dashboard. 

These functions will convert the plot object to a Panel pane.

Be sure to include any DataFrame transformation/manipulation code required along with the plotting code.

Return a Panel pane object from each function that can be used to build the dashboard.

Note: Remove any `.show()` lines from the code. We want to return the plots instead of showing them. The Panel dashboard will then display the plots.

```python

# Define Panel Visualization Functions
def housing_units_per_year():
    """Housing Units Per Year."""
    #setup data
    sfo_average = sfo_data.groupby('year').mean().dropna()
    sfo_average.sort_values(['year','housing_units'],inplace=True)
    #setup plot
    ax = sfo_average.plot(kind='bar',y='housing_units')
    ax.set_title("Average Housing Units Per Year")
    ax.set_ylim(ymin = sfo_average.housing_units.min()-4000,ymax = sfo_average.housing_units.max())
    ax.set_ylabel('Avg Housing Units')
    
    return ax
   

def average_gross_rent():
    """Average Gross Rent in San Francisco Per Year."""
    #setup data
    sfo_average.sort_values(['year','gross_rent'],inplace=True)
    #setup plot
    bx = sfo_average.plot(y='gross_rent')
    bx.set_title("San Francisco Average Gross Rent")
    bx.set_ylim(ymin = sfo_average.gross_rent.min(),ymax = sfo_average.gross_rent.max())
    bx.set_ylabel('Gross Rent')
    
    return bx
   

def average_sales_price():
    """Average Sales Price Per Year."""
    #setup data
    sfo_average.sort_values(['year','sale_price_sqr_foot'],inplace=True)
    #setup plot
    sales_ax = sfo_average.plot(y='sale_price_sqr_foot')
    sales_ax.set_title("Average Sales Price Per Year")
    sales_ax.set_ylim(ymin = sfo_average.sale_price_sqr_foot.min(),ymax = sfo_average.sale_price_sqr_foot.max())
    sales_ax.set_ylabel('Average Sales Price Per Sq Foot')
    
    return sales_ax


def average_price_by_neighborhood():
    """Average Prices by Neighborhood."""
    #setup data
    year_neighborhood_df = sfo_data.groupby(['year','neighborhood']).mean()
    #plot
    return year_neighborhood_df.sale_price_sqr_foot.hvplot.line(x='year',y='sale_price_sqr_foot',groupby='neighborhood')
    


def top_most_expensive_neighborhoods():
    """Top 10 Most Expensive Neighborhoods."""
    #setup data
    sfo_neighborhood_df = sfo_data.groupby(['neighborhood']).mean()
    price_by_neighborhood = sfo_neighborhood_df.sort_values('sale_price_sqr_foot',ascending=False)
    #setup plot
    ax = price_by_neighborhood.sale_price_sqr_foot.nlargest(10).plot.barh(y=price_by_neighborhood.index)
    ax.set_title("Neighborhoods by Sale Price")
    ax.set_xlim(xmin = price_by_neighborhood.sale_price_sqr_foot.min(),xmax = price_by_neighborhood.sale_price_sqr_foot.max())
    ax.set_xlabel('Sale Price / Sq Foot')
    ax.invert_yaxis()
   
    return ax

def parallel_coordinates():
    """Parallel Coordinates Plot."""
    #setup data
    price_by_neighborhood.sort_values(['sale_price_sqr_foot'])
    
    #setup plot
    plot = px.parallel_coordinates(price_by_neighborhood.nlargest(10,'sale_price_sqr_foot'),color='sale_price_sqr_foot',dimensions=['sale_price_sqr_foot','housing_units','gross_rent'])
    return plot

def parallel_categories():
    """Parallel Categories Plot."""
    #setup plot
    visual = px.parallel_categories(price_by_neighborhood.nlargest(10,'sale_price_sqr_foot'),
    dimensions=["neighborhood","sale_price_sqr_foot","housing_units","gross_rent"],
    color="sale_price_sqr_foot",
    color_continuous_scale=px.colors.sequential.Inferno,
    labels={
        'neighborhood' : 'San Fran Neighborhood',
        "sale_price_sqr_foot": "Sale Price / Square Foot",
        "housing_units": "Number of Housing Units",
        "gross_rent": "Gross Rent",
    },
    )
    return visual

def neighborhood_map():
    """Neighborhood Map"""
    #setup data
    file_path = Path("Data/neighborhoods_coordinates.csv")
    df_neighborhood_locations = pd.read_csv(file_path,index_col='Neighborhood')
    sales_by_nhood = sfo_data.groupby(['neighborhood']).mean()
    sales_by_nhood.index.name = 'Neighborhood'
    #Combining Neighborhood & Sales Data
    neighborhood_sales_location_df = pd.concat([sales_by_nhood,df_neighborhood_locations], axis="columns", join="inner")
    #Plot setup
    map_plot = px.scatter_mapbox(
        neighborhood_sales_location_df,
        lat="Lat",
        lon="Lon",
        size="sale_price_sqr_foot",
        color=neighborhood_sales_location_df.index,
        animation_frame='gross_rent',
        zoom=10,)

    return map_plot
```

## Panel Dashboard

In this section, you will combine all of the plots into a single dashboard view using Panel. Be creative with your dashboard design!

```python
# Create row
column = pn.Column(top_most_expensive_neighborhoods(),parallel_coordinates(),parallel_categories(),align='center')
column2 = pn.Column(average_price_by_neighborhood(),housing_units_per_year(), average_gross_rent(),average_sales_price(),align='start')
tabs = pn.Tabs(
    ("San Francisco Average Housing Data", column2),
    ("San Francisco Community Comparison Data", column),
    ("San Francisco Housing Data Map", neighborhood_map()))
```

## Serve the Panel Dashboard

```python
tabs.servable()
```
